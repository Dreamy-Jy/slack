```python
import shutil
import os
from skip import Schematic
from pathlib import Path

schematics = [
    Path("/Users/jordanethomas/Documents/Projects/keyboards/slack/hardware/v1-electronics/led_matrix.kicad_sch"),
    Path("/Users/jordanethomas/Documents/Projects/keyboards/slack/hardware/v1-electronics/controller.kicad_sch")
]

def generate_label_renames() -> dict[str, str]:
    rename = {}

    for h in [1,2]:
        # SW label changes
        c = ['','B','G','R','W','B','G','R','W','B','G','R','W','B','G','R','W']        
        for p in range (1,13):
            rename[f'SW{p}_{h}'] = f'SW{p}_{c[p]}_{h}'
        
        # CS label changes
        for p in range(1,17):
            rename[f'CS{p}_{c[p]}_{h}'] = f'CS{p}_{h}'
        
    return rename

def backup_schematics():
    for path in schematics:
        if not path.exists():
            print(f"file did not exist: {path}")
            continue
        
        shutil.copy2(path, path.with_suffix(path.suffix + ".bak"))

    print("created backups for files that exist")
        

def delete_backup_schematics():
    for path in schematics:
        backup = path.with_suffix(path.suffix + ".bak")
        if not backup.exists():
            print(f"backup file did not exist: {backup}")
            continue

        backup.unlink()
    
    print("deleted backup files that exist")
```

```python
backup_schematics()
```

```python
'''
rename targeted global labels
'''
renames = generate_label_renames()

for h in range(0,2):
    sch = Schematic(schematics[h])

    for l in sch.global_label:
        # print(l.value)
        if l.value in renames:
            l.value = renames[l.value]
    
    sch.overwrite()
```

```python
delete_backup_schematics()
```

```python
'''
fix renaming errors
'''
from collections import OrderedDict

def generate_label_fixes() -> dict[str, str]:
    fixes = [OrderedDict(), OrderedDict(),]

    for h in [1,2]:
        # SW label changes
        c = ['','B','G','R','W','B','G','R','W','B','G','R','W','B','G','R','W']        
        for p in range (12,0, -1):
            fixes[h-1][f'SW{p}'] = f'SW{p}_{c[p]}_{h}'
        
    return fixes


backup_schematics()


renames = generate_label_fixes()
# print(renames)
prefixes = renames[0].keys()
# print(prefixes)

for h in range(0,2):
    sch = Schematic(schematics[h])

    for l in sch.global_label:
        for pre in prefixes:
            if l.value.startswith(pre):
                if l.value.endswith('1'):
                    l.value = renames[0][pre]
                    # print((l.value, pre, renames[0][pre]))
                    break
                if l.value.endswith('2'):
                    l.value = renames[1][pre]
                    # print((l.value, pre, renames[1][pre]))
                    break
    
    sch.overwrite()
```

```python
delete_backup_schematics()
```
